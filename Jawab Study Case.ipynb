```python
# 1
tamu = ['Andi', 'Bayu', 'Cika', 'Dini', 'Erlin']
for nama in tamu:
  print("Halo {}, aku mengundangmu untuk makan malam".format(nama))
```

```python
# 2
tamu[2] = 'Ceyi'
for nama in tamu:
  print(f"Halo {nama}, aku mengundangmu untuk makan malam")
```

```python
# 3
tamu.insert(0, 'Farhan')
print(tamu)
tamu.insert(3, 'Gilang')
print(tamu)
tamu.append('Hanis')
print(tamu)
```

```python
for nama in tamu:
  if (nama in ['Farhan', 'Gilang', 'Hanis']):
    print("Halo {}, aku mengundangmu untuk makan malam".format(nama))
  else:
    print("Halo {}, akan ada tambahan anggota dikarenakan meja makan dapat menampung banyak orang".format(nama))
```

```python
# 4
for nama in tamu:
  print("Halo {}, maaf karena ada kesalahan reservasi sehingga aku hanya bisa mengundang 2 orang untuk makan malam".format(nama))
```

```python
tamu_diundang = ['Andi', 'Dini'] 
for nama in tamu.copy():
  if nama in tamu_diundang:
    continue
  print("Maaf {}, Aku menyesal tidak dapat mengundang kamu untuk makan malam".format(nama))
  tamu.remove(nama)
```

```python
for nama in tamu:
  print("Halo {}, kamu masih aku undang untuk makan malam".format(nama))
```

```python
del tamu[:]
print(tamu)
```
