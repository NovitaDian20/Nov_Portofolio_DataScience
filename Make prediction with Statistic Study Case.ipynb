# **Latihan (Probabilitas)**

Asosiasi  Olah Raga Nasional melakukan survei terhadap orang-orang berusia 7 tahun atau lebih tua tentang partisipasi dalam kegiatan olahraga (Abstrak Statistik Amerika Serikat: 2002).
Jumlah penduduk pada kelompok umur ini dilaporkan sebanyak 248,5 juta, dengan 120,9 juta laki-laki dan 127,6 juta perempuan. Jumlah peserta untuk lima aktivitas olahraga teratas adalah \\

|Activity| Male |Female|
|----|----|-----|
Bicycle riding |22.2| 21.0
Camping |25.6 |24.3
Exercise walking| 28.7 |57.7
Exercising with equipment |20.4 |24.4
Swimming |26.4| 34.4

a. Untuk wanita yang dipilih secara acak, perkirakan probabilitas partisipasi di masing-masing 5 kegiatan olahraga? \\
b. Untuk orang yang dipilih secara acak, berapa probabilitas orang tersebut berpartisipasi olahraga excercise walking?

```python
# Jawaban A
# Menerapkan dictionary untuk membuat dataframe
df_probabilitas = pd.DataFrame({'Activity': ['Bicycle riding', 'Camping', 'Exercise walking', 'Exercising with equipment', 'Swimming'],
                                'Male': [22.2, 25.6, 28.7, 20.4, 26.4],
                                'Female': [21.0, 24.3, 57.7, 24.4, 34.4]})

# Display dataframe
df_probabilitas
```

**Referensi Jawaban 1**

```python
prob_female = df_probabilitas.groupby('Activity')['Female'].sum() / sum(df_probabilitas['Female'])

print("Probabilitas wanita yang dipilih secara acak untuk setiap kegiatan olahraga adalah")
print(prob_female)
```

```python
# Jawaban B
df_probabilitas['Participants'] = df_probabilitas['Female'] + df_probabilitas['Male']

prob_choose = df_probabilitas.groupby('Activity')['Participants'].sum().loc['Exercise walking'] / sum(df_probabilitas['Participants'])

print("Probabilitas orang yang dipilih secara acak untuk kegiatan olahraga Exercise walking adalah " + str(prob_choose))
```

**Referensi Jawaban 2**

```python
# Set activity sebagai index
df_probabilitas_new = df_probabilitas.set_index('Activity')
df_probabilitas_new
```

```python
df_probability_new
```

```python
# Jawaban A
prob_female = df_probabilitas_new['Female'] / df_probabilitas_new['Female'].sum()

print("Probabilitas wanita yang dipilih secara acak untuk setiap kegiatan olahraga adalah")
print(prob_female)
```

```python
# Jawaban B
df_tot = df_probabilitas_new['Female'] + df_probabilitas_new['Male']

prob_choose = df_tot['Exercise walking'] / df_tot.sum()
print("Probabilitas orang yang dipilih secara acak untuk kegiatan olahraga excercise walking adalah " + str(prob_choose))
```

# **Latihan (Statistika Deskriptif)**

Untuk lebih memahami materi statistika deskriptif, silakan untuk mencoba latihan ini

Untuk periode 2001â€“2008, Bristol-Myers Squibb
Company, Inc. melaporkan jumlah (dalam milyaran
dolar) untuk (1) penjualan bersih dan (2) iklan dan
promosi produk \\

| Tahun | Net sales | Iklan |
|----|---|----|
|2001 | 16.612 |1.201|
|2002 |16.208 |1.143|
|2003 |18.653 |1.416|
|2004 |19.380 |1.411|
|2005 |19.207 |1.476|
|2006 |16.208 |1.304|
|2007 |18.193 |1.415|
|2008 |20.597| 1.550|

(a) Tunjukkan berapa rata-rata net sales dan iklan selama 9 tahun? \\
(b) Berapa variasi yang didapatkan untuk net sales dan iklan selama 9 tahun? \\


```python
# Jawaban A
# Menerapkan dictionary untuk membuat dataframe
df_stat = pd.DataFrame({'Tahun': [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008],
                        'Net sales': [16.612, 16.208, 18.653, 19.380, 19.207, 16.208, 18.193, 20.597],
                        'Iklan': [1.201, 1.143, 1.416, 1.411, 1.476, 1.304, 1.415, 1.550]})

# Display dataframe
df_stat
```

```python
# Rata-rata net sales dan iklan selama 9 tahun
# Mendapatkan mean pada kolom net sales
mean_net_sales = df_stat['Net sales'].mean()

# Mendapatkan mean pada kolom Iklan
mean_iklan= df_stat['Iklan'].mean()

# Display hasil
print("Nilai rata-rata net sales selama 9 tahun adalah " + str(mean_net_sales))
print("Nilai rata-rata iklas selama 9 tahun adalah " + str(mean_iklan))
```

**Menggunakan library statistics**

```python
# Jawaban B
import statistics

# Mendapatkan varian dari masing-masing kolom
print("Varians dari net sales selama 9 tahun adalah :", statistics.variance(df_stat['Net sales']))
print("Varians dari iklan selama 9 tahun adalah :", statistics.variance(df_stat['Net sales']))
```

**Tidak menggunakan library statistics**

```python
# Mendapatkan varian dari masing-masing kolom
print("Varians dari net sales selama 9 tahun adalah :", df_stat['Net sales'].var())
print("Varians dari iklan selama 9 tahun adalah :", df_stat['Net sales'].var())
```

# Latihan (Statistika Deskriptif di bisnis)

Berdasarkan data transaksi dari coffee shop diatas. Carilah persebaran umur berdasarkan mean, dan varian dari customer tersebut. Serta dapatkan modus usia dari customers.

```python
# Panggil kembali data coffee shop disini

# Read dataset
df_cust = pd.read_csv(path + 'customer.csv')
df_rec = pd.read_csv(path + 'reciepts.csv')

# Join columns (Merging.. Left Join..) 
df_trx = pd.merge(df_cust, df_rec, left_on = 'customer_id', right_on = 'customer_id', how = 'left')

df_trx.head()
```

```python
# Mencari deskripsi data
df_trx.info()
```

```python
# Merubah kolom birthdate kedalam format datetime
df_trx['birthdate'] = pd.to_datetime(df_trx['birthdate'])

# Cek perubahan format data
df_trx.dtypes
```

```python
# Mencari usia berdasarkan data birthdate
import datetime as dt

df_trx['day'] = (dt.datetime.now() - df_trx['birthdate']).dt.days

# Untuk mendapatkan tahun maka hari akan dibagi dengan total hari dalam setahun
df_trx['usia'] = df_trx['day'] / 365

# Menampilkan perubahan yang dilakukan
df_trx.head(6)
```

```python
# Mencari statistika deskriptif dari kolom umur
df_trx['usia'].describe()
```
