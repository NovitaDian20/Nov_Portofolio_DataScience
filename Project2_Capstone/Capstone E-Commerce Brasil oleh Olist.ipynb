## EXPLORATORY DATA

```python
#dengan bantuan statistik/pencilan/outliers
sns.boxplot(x='freight_value', data=data)
```

```python
Q1=data['freight_value'].quantile(0.25)
Q3=data['freight_value'].quantile(0.75)
IQR=Q3-Q1
Lb=Q1-(IQR*1.5)
Ub=Q3+(IQR*1.5)
print('Batas atas &Batas bawah :', Lb, Ub)
```

```python
data_new=data[(data['freight_value']<Ub)&(data['freight_value']>Lb)]
data_new.head()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='freight_value', data=data_new)
```

```python
sns.boxplot(x='price', data=data)
```

```python
Q1=data['price'].quantile(0.25)
Q3=data['price'].quantile(0.75)
IQR=Q3-Q1
Lb=Q1-(IQR*1.5)
Ub=Q3+(IQR*1.5)
print('Batas atas &Batas bawah :', Lb, Ub)
```

```python
data_new1=data[(data['price']<Ub)&(data['price']>Lb)]
data_new1.head()
```
